

\#1 What is a random variable in probability theory?
- A random variable is a numerical representation of outcomes from a random experiment.
It maps outcomes from a sample space to real numbers.
There are two main types: discrete and continuous.

---

\#2 What are the types of random variables?
- The two types of random variables are discrete and continuous.
Discrete variables take on specific, countable values, while continuous variables can take any value within a range.
Each is used in different kinds of probability models.

---

\#3 What is the difference between discrete and continuous distributions?
- Discrete distributions deal with countable outcomes like dice rolls or coin tosses.
Continuous distributions involve measurable quantities like height or temperature.
They use PMF and PDF respectively to express probabilities.

---

\#4 What are probability distribution functions (PDF)?
- A PDF represents the probability density of a continuous random variable.
It shows the likelihood of the variable falling within a particular range.
The total area under a PDF curve equals 1.

---

\#5 How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
- A CDF gives the probability that a variable is less than or equal to a certain value.
PDF shows the likelihood over an interval, while CDF accumulates probabilities.
CDF is always non-decreasing and ranges from 0 to 1.

---

\#6 What is a discrete uniform distribution?
- A discrete uniform distribution assigns equal probability to all outcomes.
For example, in rolling a fair die, each face has an equal chance of appearing.
It is used when every result is equally likely.

---

\#7 What are the key properties of a Bernoulli distribution?
- A Bernoulli distribution models a binary outcome — success (1) or failure (0).
It has a single trial and a probability `p` of success.
Mean is `p`, and variance is `p(1-p)`.

---

\#8 What is the binomial distribution, and how is it used in probability?
- The binomial distribution models the number of successes in `n` independent Bernoulli trials.
Each trial has the same probability `p` of success.
It’s used in situations like coin tosses or quality control.

---

\#9 What is the Poisson distribution and where is it applied?
- The Poisson distribution models the number of times an event occurs in a fixed interval.
It assumes events occur independently and at a constant average rate.
Common in modeling arrivals (e.g., calls to a help desk).

---

\#10 What is a continuous uniform distribution?
- - A continuous uniform distribution assigns equal probability density to all values in an interval.
It is defined over a range `[a, b]`, where every value is equally likely.
The PDF is `1 / (b - a)` for `a ≤ x ≤ b`.

---

\#11 What are the characteristics of a normal distribution?
- A normal distribution is bell-shaped and symmetric about the mean.
It is defined by its mean (μ) and standard deviation (σ).
Most real-world phenomena follow or approximate this distribution.

---

\#12 What is the standard normal distribution, and why is it important?
- The standard normal distribution is a normal distribution with mean 0 and standard deviation 1.
It is used to standardize scores (z-scores) across datasets.
It forms the basis of many statistical tests and confidence intervals.

---

\#13 What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- The CLT states that the sampling distribution of the mean approaches normality as sample size increases.
It applies regardless of the population's original distribution.
It justifies using normal-based methods for inference.

---

\#14 How does the Central Limit Theorem relate to the normal distribution?
- The CLT explains why sample means tend to form a normal distribution even from non-normal data.
As sample size increases, the distribution of sample means becomes approximately normal.
This enables use of z-scores and other tools.

---

\#15 What is the application of Z statistics in hypothesis testing?
- Z-statistics help compare sample results to a known population.
They determine how far a sample mean is from the population mean under the null hypothesis.
It’s widely used in large-sample hypothesis testing.

---

\#16 How do you calculate a Z-score, and what does it represent?
- Z = (X - μ) / σ, where X is a data point, μ is mean, and σ is standard deviation.
It tells how many standard deviations a value is from the mean.
Z-scores help identify outliers and standardize data.

---

\#17 What are point estimates and interval estimates in statistics?
- A point estimate provides a single value as an estimate of a population parameter.
An interval estimate gives a range (like confidence intervals) within which the parameter lies.
Interval estimates offer more information and account for uncertainty.

---

\#18 What is the significance of confidence intervals in statistical analysis?
- Confidence intervals estimate a population parameter with a given level of confidence (e.g., 95%).
They provide a range of plausible values instead of a single estimate.
They help assess the reliability of statistical results.

---

\#19 What is the relationship between a Z-score and a confidence interval?
- Z-scores define the margins of a confidence interval around a sample mean.
For a 95% CI, the z-score is typically ±1.96.
Higher confidence levels correspond to wider intervals.

---

\#20 How are Z-scores used to compare different distributions?
- Z-scores standardize different datasets to a common scale.
They allow comparison of values from different distributions regardless of units.
This is useful in cross-group or cross-variable analysis.

---

\#21 What are the assumptions for applying the Central Limit Theorem?
- Samples must be independent and identically distributed.
The sample size should be sufficiently large (usually n ≥ 30).
If the population is normal, even small samples work.

---

\#22 What is the concept of expected value in a probability distribution?
- Expected value is the long-term average outcome of a random variable.
It is calculated by summing the products of each value and its probability.
It gives the "center" or mean of the distribution.

---

\#23 How does a probability distribution relate to the expected outcome of a random variable?
- The distribution defines all possible outcomes and their probabilities.
The expected outcome (mean) is a weighted average based on this distribution.
It’s a key concept in making predictions from data.
